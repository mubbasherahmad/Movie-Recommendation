```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['production_companies']=movies['production_companies'].apply(convert)
```

```python
movies.production_companies
```

```python
def convert2(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter=counter+1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
movies.cast
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.crew.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x ])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x ])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x ])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x ])
movies['production_companies']=movies['production_companies'].apply(lambda x:[i.replace(" ","")for i in x ])
```

```python
movies.head()
```

```python
movies['tags']= movies['overview']+movies['genres']+movies['cast']+movies['crew']+movies['production_companies']
+movies['keywords']
```

```python
movies.tags.head(1).values
```

```python
fmovies=movies[["movie_id","title","tags"]]
```

```python
fmovies.head()
```

```python
fmovies['tags']=fmovies['tags'].apply(lambda x:" ".join(x))
```

```python
fmovies.tags
```

```python
fmovies.tags.head(1).values
```

```python
fmovies['tags']=fmovies['tags'].apply(lambda x:x.lower())
```

```python
fmovies.head(1)
```

```python
fmovies.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
fmovies['tags']=fmovies['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,stop_words='english')
```

```python
vectors=cv.fit_transform(fmovies['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[199])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(movie):
    movie_index=fmovies[fmovies['title'] == movie].index[0]
    distance = similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(fmovies.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
